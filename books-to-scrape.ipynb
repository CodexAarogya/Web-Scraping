### Scraping book's name and prices having price less than £30

```python
from bs4 import BeautifulSoup
import requests
import html.parser, lxml
```

```python
url = "https://books.toscrape.com/"   # legal scraping site
response = requests.get(url)
response.text                         # returns HTML from the site
```

```python
data = BeautifulSoup(response.text, "html.parser")   # parsed HTML
data
```

```python
product_price = data.find_all("div", class_="product_price")
for prices in product_price:
    # print(prices)
    price = prices.find('p', class_="price_color")
    if float(price.get_text(strip=True)[2:]) < 30:
        title = prices.parent.find('h3').find('a')['title']
        print(title)
        print(price.get_text(strip=True)[1:])
```
